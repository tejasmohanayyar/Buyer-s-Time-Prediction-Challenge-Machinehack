###### Creating date columns

```python
train['date'] = train['date'].astype('datetime64')
train['dayofmonth'] = train['date'].dt.day
train['weekofyear'] = train['date'].dt.week
train['dayofweek'] = train['date'].dt.dayofweek
train['year'] = train['date'].dt.year
train['month'] = train['date'].dt.month
train['dayofyear'] = train['date'].dt.dayofyear
```

```python
train['device'] = [i.split('-')[0].strip() for i in train['device_details']]
train['platform'] = [i.split('-')[1].strip() for i in train['device_details']]
```

```python
for features in ['device_details', 'device', 'platform']:
    plt.figure(figsize = (15,6))
    plt.xticks(rotation = 90)
    sns.boxplot(features,'time_spent',data = train)
```

```python
train['apc'] = train['added_in_cart']+train['purchased']+train['checked_out']
```

```python
from sklearn.feature_selection import SelectFromModel
from sklearn.linear_model import Lasso,Ridge,LinearRegression
from sklearn.preprocessing import MinMaxScaler,StandardScaler
from sklearn.model_selection import train_test_split
import category_encoders as ce
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
model = ols('time_spent~apc',data = train).fit()
sm.stats.anova_lm(model)[sm.stats.anova_lm(model).loc[:,'PR(>F)']<0.05]
```

```python
X = train.drop('time_spent',axis = 1)
y = train['time_spent']
cbe = ce.CatBoostEncoder(cols = ['device','platform','apc'])
train = cbe.fit_transform(X,y)
```

```python
train.drop(['session_id','session_number','client_agent','date','year','device_details','added_in_cart','checked_out','purchased','dayofmonth','weekofyear','dayofweek','month'],axis = 1,inplace = True)
```

```python
train
```

```python
ss = MinMaxScaler()
for i in ['dayofyear','device','platform','apc']:
    train[i] = ss.fit_transform(np.array(train[i]).reshape(-1,1))
```

```python
X = np.array(train)
```

```python
y = ss.fit_transform(np.array(y).reshape(-1,1))
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,y,test_size = 0.1,random_state = 6)
```

```python
x_train.shape
```

```python
from sklearn.linear_model import LinearRegression,Lasso,Ridge,ElasticNet
from sklearn.metrics import mean_squared_log_error
```

```python
msle_scores = {}
```

```python
lr = LinearRegression()
lr.fit(x_train,y_train)
lr.score(x_test,y_test)
y_pred = lr.predict(x_test)

msle_scores.setdefault('linear_reg',mean_squared_log_error(y_test,y_pred))
sns.regplot(y_test,y_pred)
mean_squared_log_error(y_test,y_pred),lr.score(x_test,y_test)
```

# Ridge

```python
lr = Ridge()
lr.fit(x_train,y_train)
lr.score(x_test,y_test)
y_pred = lr.predict(x_test)

msle_scores.setdefault('Ridge_reg',mean_squared_log_error(y_test,y_pred))
sns.regplot(y_test,y_pred)
mean_squared_log_error(y_test,y_pred),lr.score(x_test,y_test)
```

# Lasso Regression

```python
lr = Lasso(alpha=0.05)
lr.fit(x_train,y_train)
lr.score(x_test,y_test)
y_pred = lr.predict(x_test)

msle_scores.setdefault('Lasso_reg',mean_squared_log_error(y_test,y_pred))
sns.regplot(y_test,y_pred)
mean_squared_log_error(y_test,y_pred),lr.score(x_test,y_test)
```

# Elastic Net

```python
lr = ElasticNet(alpha=0.01)
lr.fit(x_train,y_train)
lr.score(x_test,y_test)
y_pred = lr.predict(x_test)

msle_scores.setdefault('Lasso_reg',mean_squared_log_error(y_test,y_pred))
sns.regplot(y_test,y_pred)
mean_squared_log_error(y_test,y_pred),lr.score(x_test,y_test)
```

```python
Observed = y_pred
Expected = y_test.reshape(1,-1)

chi = abs(Observed-Expected)
```
