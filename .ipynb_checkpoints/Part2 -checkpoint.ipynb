```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.linear_model import LinearRegression,Ridge,Lasso
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.metrics import r2_score,mean_squared_log_error
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import BaggingRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.neighbors import KNeighborsRegressor
import xgboost as xgb
```

```python
train = pd.read_csv('data/Train.csv')
test = pd.read_csv('data/Test.csv')
```

```python
train.head()
```

```python
train.drop(['session_id','session_number','client_agent'],axis = 1,inplace = True)
test.drop(['session_id','session_number','client_agent'],axis = 1,inplace = True)
```

```python
train.shape,test.shape
```

```python
train.info()
```

```python
train['date'] = train['date'].astype('datetime64')
test['date'] = test['date'].astype('datetime64')
```

```python
train['day'] = [i.day for i in train['date']]
train['month'] = [i.month for i in train['date']]
train['year'] = [i.year for i in train['date']]

test['day'] = [i.day for i in test['date']]
test['month'] = [i.month for i in test['date']]
test['year'] = [i.year for i in test['date']]
```

```python
train['week'] = [i.week for i in train['date']]
test['week'] = [i.week for i in test['date']]
```

```python
train.drop('date',axis = 1,inplace = True)
test.drop('date',axis = 1,inplace = True)
```

```python
train.shape,test.shape
```

```python
dev_det_tr = pd.get_dummies(train['device_details'])
dev_det_test = pd.get_dummies(test['device_details'])
```

```python
train.shape,test.shape
```

```python
dev_det_tr.drop(np.setdiff1d(dev_det_tr.columns,dev_det_test.columns)[0],axis = 1,inplace = True)
```

```python
dev_det_tr.shape,dev_det_test.shape
```

```python
train = pd.concat((train,dev_det_tr),axis = 1)
test = pd.concat((test,dev_det_test),axis = 1)
```

```python
train.shape,test.shape
```

```python
test.drop('device_details',axis = 1,inplace = True)
train.drop('device_details',axis = 1,inplace = True)
```

```python
plt.figure(figsize = (30,20))
sns.heatmap(train.corr(),annot = True)
plt.show()
```

```python
train['sum'] = train['purchased']+train['added_in_cart']+train['checked_out']
test['sum'] = test['purchased']+test['added_in_cart']+test['checked_out']
```

```python
train.drop(['purchased','added_in_cart','checked_out','month'],axis = 1,inplace = True)
test.drop(['purchased','added_in_cart','checked_out','month'],axis = 1,inplace = True)
```

```python
train.columns
```

```python
train.shape,test.shape
```

# Regression

```python
# train.drop('year',axis = 1,inplace = True)
# test.drop('year',axis = 1,inplace = True)
```

```python
ss = StandardScaler()
X = train.drop('time_spent',axis = 1)
X_scaled = ss.fit_transform(X)
y = train.time_spent
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif = pd.DataFrame()
vif["vif"] = [variance_inflation_factor(X_scaled,i) for i in range(X_scaled.shape[1])]
vif["Features"] = X.columns
#let's check the values
vif
```

###### Seems to me that iPhone ioss shows a high inflation factor. We will also be dropping the year column. Lets remove these columns and see what happens..

```python
train.drop(['iPhone - iOS','year','day'],axis = 1,inplace = True)
test.drop(['iPhone - iOS','year','day'],axis = 1,inplace = True)
```

```python
ss = StandardScaler()
X = train.drop('time_spent',axis = 1)
X_scaled = ss.fit_transform(X)
y = train.time_spent
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif = pd.DataFrame()
vif["vif"] = [variance_inflation_factor(X_scaled,i) for i in range(X_scaled.shape[1])]
vif["Features"] = X.columns
#let's check the values
vif
```

###### WOW! All our VIF's have arrived close to 1 which is acceptable.

```python
x_train,x_test,y_train,y_test = train_test_split(X_scaled,y,test_size = 0.2,random_state = 6)
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
lr = LinearRegression()
lr.fit(x_train,y_train)
lr.score(x_test,y_test)
```

```python
y_pred = abs(lr.predict(x_test))
```

```python
def scoring(y_test,y_pred):
    sns.regplot(y_test,y_pred)
    return np.sqrt(mean_squared_log_error(y_test,y_pred))
scoring(y_test,y_pred)
```

```python
sns.regplot(y_test,y_pred)
plt.show()
```

```python
test_scaled = ss.fit_transform(test)
y_pred = lr.predict(test_scaled)
```

```python
sub = pd.DataFrame(y_pred,columns = ['time_spent'])
sub.to_csv('Submissions/LinearRegression.csv',index = False)
```
