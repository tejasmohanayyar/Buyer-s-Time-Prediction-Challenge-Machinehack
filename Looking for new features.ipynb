```python
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
import pandas as pd
import numpy as np
import scipy.stats as stats
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
train = pd.read_csv('Data/Train.csv')
test = pd.read_csv('Data/Test.csv')
```

```python
train['date'] = train['date'].astype('datetime64')
```

```python
train['month'] = train['date'].dt.month
```

```python
sns.jointplot('month','time_spent',data = train)
```

```python
test['date'] = test['date'].astype('datetime64')
```

```python
test['month'] = test['date'].dt.month
```

```python
test.groupby('month')['month'].count()
```

```python
train['device'] = [i.split('-')[0].strip() for i in train['device_details']]
test['device'] = [i.split('-')[0].strip() for i in test['device_details']]

train['platform'] = [i.split('-')[1].strip() for i in train['device_details']]
test['platform'] = [i.split('-')[1].strip() for i in test['device_details']]
```

```python
train.drop(['session_number','session_id','date','device_details'],axis = 1,inplace = True)
test.drop(['session_number','session_id','date','device_details'],axis = 1,inplace = True)
```
